# Importing Libraries

```python
import pandas as pd
import datasist as ds
import numpy as np
```

# Reading Dataset

```python
train_df=pd.read_csv('C:/Users/Apar/Downloads/train_data.csv')
```

# Handle missing values (drop/fill)

```python
#CHECKING MISSING VALUES
ds.structdata.display_missing(train_df)
```

```python
#drop the column with 7 percent missing values 
new_train_df = ds.feature_engineering.drop_missing(train_df,  
                                                    percent=7.0)
ds.structdata.display_missing(new_train_df)
```

```python
# Filling Categorical Features having missing values
df = ds.feature_engineering.fill_missing_cats(new_train_df)
ds.structdata.display_missing(df)
```

```python
# Filling Numerical Featres having missing values
df = ds.feature_engineering.fill_missing_num(df)
ds.structdata.display_missing(df)
```
